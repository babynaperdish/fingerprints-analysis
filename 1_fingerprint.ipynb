Для работы сканера ZK9500 необходимо сначала установить драйвер (отправил архив в телеграме). Все запросы к сканеру отправляются с помощью библиотеки `pyzkfp` (<a href='https://github.com/aqasemi/pyzkfp'>ссылка</a>)

Импортируем необходимые библиотеки (если какой-либо библитеки нет, то предварительно `pip install ...`):

```python
from matplotlib.pyplot import imshow
import numpy as np
from PIL import Image
from pyzkfp import ZKFP2
```

```python
zkfp2 = ZKFP2()
zkfp2.Init()

device_count = zkfp2.GetDeviceCount()
print(f"{device_count} devices found")
zkfp2.OpenDevice(0)
```

```python
while True:
 capture = zkfp2.AcquireFingerprint()
 if capture:
     tmp, img = capture
     Image.frombytes("L", (300, 375), img).save("finger_images/finger.png", "PNG")
     imshow(Image.frombytes("L", (300, 375), img), cmap="gray", vmin=0, vmax=255)
     break
```

```python
zkfp2.Terminate()
```
